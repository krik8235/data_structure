### Pathfinding Algorithms

- process of finding the shortest path between two points in a graph.


#### Two key pathfinding algorithms:



### 1. Dijkstra's Algorithm:
- Finds the shortest path from a source vertex to all other vertices in a graph.
- Works by iteratively selecting the vertex with the minimum distance and exploring its neighbors.
- Time complexity: O((V+E) log V) - (ref. #17)



### 2. A* Algorithm:
- A modification of Dijkstra's algorithm that uses a heuristic function to estimate the distance to the goal

- a greedy algo using heuristic function to estimate the cost of reaching the goal from a given node
- always expands the node with the lowest estimated total cost (f-score)
    - f-score = the cost to reach the node (g-score) + the estimated cost to the goal (h-score)

- The greedy algo approach guides the search towards the goal more efficiently than Dijkstra's algo in many cases

- Time complexity: O(bd)
    - b: the branching factor
    - d: depth of the goal

    - Cf. Dijkstra: O((V+E) log V) - influenced by the vertices and edges that we will not visit

- Dijkstra's algorithm is particularly useful when we want to find the shortest paths from a single source node to all other nodes in the graph. In contrast, A* is more efficient when we have a specific target node and want to find the shortest path to that node. Â  


#### Key Points:
- Dijkstra's algorithm can be inefficient for single-source, single-destination problems as it explores all vertices.
- A* algorithm is more efficient by focusing on paths that are likely to lead to the goal.
- The choice of heuristic function can significantly impact the performance of A*.
- Both algorithms are fundamental tools in computer science and have applications in various fields.
